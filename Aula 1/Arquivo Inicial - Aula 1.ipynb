# Automação de Sistemas e Processos com Python

### Desafio:

Todos os dias, o nosso sistema atualiza as vendas do dia anterior.
O seu trabalho diário, como analista, é enviar um e-mail para a diretoria, assim que começar a trabalhar, com o faturamento e a quantidade de produtos vendidos no dia anterior

E-mail da diretoria: seugmail+diretoria@gmail.com<br>
Local onde o sistema disponibiliza as vendas do dia anterior: https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing

Para resolver isso, vamos usar o pyautogui, uma biblioteca de automação de comandos do mouse e do teclado

```python

# vamos precisar instalar
#pyautogui -> automações
#pandas -> dados
#openpyxl -> trabalhar com bases de dados em Excel (.xlsx)
# para instalar: abrir o terminal e rodar: pip install pandas pyautogui openpyxl

# pyautogui.click -> clicar com o mause
# pyautogui -> escrever um texto
# pyautogui -> aperta 1 tecla
# pyautogui -> combinação de teclas
```

```python
import pyautogui
import time

pyautogui.PAUSE = 1

# Passo a passo do desafio
# Passo 1: Entrar no sistema da empresa (link do drive)
# abrir o Chrome
pyautogui.press("win")
pyautogui.write("chrome")
pyautogui.press("enter")

pyautogui.click(x=1068, y=468)

time.sleep(1.5)
link = "https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga"
# digitei o link do sistema
pyautogui.write(link)
# apertei enter e esperei
pyautogui.press("enter")
time.sleep(3)
```

```python
import pyautogui
import time
# entrar na pasta do arquivo
pyautogui.click(x=564, y=425, clicks=2)

# Passo 2: Navegar no sistema para encontrar a base de dados
pyautogui.click(x=558, y=415)
time.sleep(3)
# Passo 3: Exportar a base de dados (baixar o arquivo)
pyautogui.click(x=669, y=419)
time.sleep(3)
pyautogui.click(x=1656, y=419)
```

### Vamos agora ler o arquivo baixado para pegar os indicadores

- Faturamento
- Quantidade de Produtos

```python
import pandas

# Passo 4: Calcular os indicadores (faturamento e quantidade de produtos vendidos)
# abrir a base de dados
caminho = r"C:\Users\user\Downloads\Vendas - Dez.xlsx"
tabela = pandas.read_excel(caminho)

# ver as informações da base de dados
display(tabela)

# somar o faturamento de todos os produtos = somar a coluna de valor final
faturamento = tabela["Valor Final"].sum()

# somar a quaantidade de produtos = somar a coluna de quantidade
qtde_produtos = tabela["Quantidade"].sum()

print(faturamento)
print(qtde_produtos)
```

### Vamos agora enviar um e-mail pelo gmail

```python
import pyautogui
import time
import pyperclip

# Passo 5: Enviar as informações por email
# abrir uma nova aba
time.sleep(1)
pyautogui.hotkey("ctrl", "t")

# entrar no email
pyautogui.write("http://mail.google.com/")
pyautogui.press("enter")
time.sleep(5)

# clicar no botão escrever
pyautogui.click(x=112, y=265)
time.sleep(4)

# para quem enviar o email
pyautogui.write("alexpernaoficial@gmail.com")
pyautogui.press("tab") # selecionar email
pyautogui.press("tab") # passar para o assunto

# qual o asssunto
pyperclip.copy("Relatório de vendas")

pyautogui.hotkey("ctrl", "v")
pyautogui.press("tab") # passar para corpo do email

# qual o corpo do email

texto = f"""
Prezado,

Segue o relatório de vendas de hoje.

Faturamento: R${faturamento:,.2f}
Quantidade de produtos vendidos: {qtde_produtos:,}

Qualquer coisa estou à disposição.
Abs,
Alex Perna
"""
pyperclip.copy(texto)
pyautogui.hotkey("ctrl", "v")

# enviar o email
pyautogui.click(x=1196, y=1000)
```
